```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")
df.head()
```

```python
# Bar: count of survivors
survived_count = df['Survived'].value_counts() 
print(survived_count)
```

```python
# Line: average age per passenger class
avg_age_class = df.groupby('Pclass')['Age'].mean()
print(avg_age_class)
```

```python
plt.figure(figsize=(6,4))
plt.subplot(1,3,1)
plt.bar(survived_count.index, survived_count.values, color='green')
plt.title("Survival Count")
plt.xlabel("0 = Died, 1 = Survived")
plt.ylabel("Number of Passengers")
```

```python
plt.figure(figsize=(6,4))
plt.subplot(1,3,2)
plt.plot(avg_age_class.index, avg_age_class.values, marker='o', color='blue', linewidth=2)
plt.title("Average Age per Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Age")
```
