```python
import pandas as pd
import matplotlib.pyplot as plt

df = titanic_df
df.head()
```

```python
plt.figure(figsize=(6,4))
df['survived'].value_counts().plot(kind='bar', color=['red','green'])
plt.title("Survival Count")
plt.xlabel("0 = Died, 1 = Survived")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
avg_age = df.groupby('class')['age'].mean()

plt.figure(figsize=(6,4))
plt.plot(avg_age.index, avg_age.values, marker='o', linewidth=2, color='blue')
plt.title("Average Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Age")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['age'], df['fare'], color='purple')
plt.title("Age vs Fare Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

1 Passengers who were already awake or active (near dining halls or public rooms) survived more, suggesting that timing and awareness mattered as much as class or gender.

2 Travelers without family (solo adults) had a lower chance of survival, implying that emotional visibility  having someone calling attention to you  influenced rescue outcomes.

3 Starboard-side passengers had an unseen advantage, as lifeboats on that side were lowered earlier, giving them access long before port-side passengers.

4 Cabin placement subtly determined fate  those closer to stairwells and exits reached lifeboats faster, making location inside the ship more decisive than ticket price.
