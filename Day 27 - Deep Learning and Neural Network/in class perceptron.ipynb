```python
# =========================
# Contoh cara kira Perceptron Manual Example
# =========================

# ----- Inputs (0 or 1) -----
x1 = int(input("Artist is good? (0/1): "))
x2 = int(input("Weather is good? (0/1): "))
x3 = int(input("Friend will come? (0/1): "))
x4 = int(input("Food is served? (0/1): "))
x5 = int(input("Juice is served? (0/1): "))

# ----- Weights -----
w1 = 0.7
w2 = 0.2
w3 = 0.6
w4 = 0.3
w5 = 0.4

# ----- Bias -----
b = -1.0   # threshold keputusan

# ----- Weighted Sum -----
z = (x1*w1) + (x2*w2) + (x3*w3) + (x4*w4) + (x5*w5) + b

# ----- Activation Function (Step) -----
if z >= 0:
    y = 1
else:
    y = 0

# ----- Output -----
print("\n--- RESULT ---")
print("Weighted Sum (z):", z)

if y == 1:
    print("Decision: YES (Go / Event happens)")
else:
    print("Decision: NO (Do not go / Event cancelled)")
```
