```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 200) # Define x

y = 2*x + 3   # Equation straight line y = 2x + 3
plt.plot(x, y, linewidth=3, label="y = 2x + 3")   # Plot line
plt.axhline(0, color='black', linewidth=1)   # horizontal axis (y=0)
plt.axvline(0, color='black', linewidth=1)   # vertical axis (x=0)
```

```python
#Seaborn Version
import seaborn as sns

x = np.linspace(-10, 10, 200) # Data
y = 2*x + 3

sns.set_style("whitegrid")   # Setup style

sns.lineplot(x=x, y=y, linewidth=3, label="y = 2x + 3")  # Plot dengan Seaborn

# Tambah paksi Cartesian
plt.axhline(0, color='black', linewidth=1)   # horizontal axis
plt.axvline(0, color='black', linewidth=1)   # vertical axis
```

```python
#Seaborn Version 2
import seaborn as sns

x = np.linspace(-10, 10, 200) # Data
y = 2*x + 3

sns.set_style("whitegrid")   # Setup style

sns.lineplot(x=x, y=y, linewidth=3, label="y = 2x + 3")  # Plot dengan Seaborn

# Tambah paksi Cartesian
plt.axhline(0, color='black', linewidth=1)   # horizontal axis
plt.axvline(0, color='black', linewidth=1)   # vertical axis

# Labels & Title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Straight Line (Seaborn): y = 2x + 3")
plt.legend()

plt.show()
```

```python
#PARABOLA (Quadratic)  y = x #try different graph
sns.set_style("whitegrid")

x = np.linspace(-10, 10, 400)
y = x**2

sns.lineplot(x=x, y=y, linewidth=3)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.title("Parabola: y = x^2")
plt.show()
```

```python
#CUBIC CURVE  y = x
x = np.linspace(-10, 10, 400)
y = x**3

sns.lineplot(x=x, y=y, linewidth=3)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.title("Cubic Curve: y = x^3")
plt.show()
```

```python
#EXPONENTIAL  y = e^x
x = np.linspace(-6, 6, 1000)
y = np.exp(x)

sns.lineplot(x=x, y=y, linewidth=3)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.title("Exponential Curve: y = e^x")
plt.show()
```

```python
#LOGARITHMIC  y = ln(x)
x = np.linspace(0.1, 10, 400)
y = np.log(x)

sns.lineplot(x=x, y=y, linewidth=3)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.title("Logarithmic Curve: y = ln(x)")
plt.show()
```

```python
#SINE WAVE  y = sin(x)
x = np.linspace(-10, 10, 400)
y = np.sin(x)

sns.lineplot(x=x, y=y, linewidth=3)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.title("Sine Wave: y = sin(x)")
plt.show()
```

```python
#ABSOLUTE VALUE  y = |x|
x = np.linspace(-10, 10, 400)
y = np.abs(x)

sns.lineplot(x=x, y=y, linewidth=3)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.title("Absolute Value: y = |x|")
plt.show()
```

```python
#HYPERBOLA  y = 1/x
x = np.linspace(-10, 10, 400)
x = x[x != 0]   # buang 0 sebab tak boleh bahagi
y = 1/x

sns.lineplot(x=x, y=y, linewidth=3)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.title("Hyperbola: y = 1/x")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

x = np.linspace(-2.5, 2.5, 600)

# W tajam, tidak flat
y = x**4 - 3*(x**2) + 0.5*x

sns.lineplot(x=x, y=y, linewidth=3)

plt.axhline(0, color='black')
plt.axvline(0, color='black')

plt.title("W Tajam (Tidak Flat): 0.2x^6 - 0.5x^4 + 2x")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

x = np.linspace(-40, 40, 1000)

y = (x**4) - 250*(x**2) +(x**3) # TRUE deep W shape

plt.plot(x, y, linewidth=3)

plt.axhline(0, color='green', linewidth=2)
plt.axvline(0, color='purple', linewidth=2)

plt.grid(True)
plt.title("TRUE DEEP W SHAPE (Degree 8 Polynomial)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
