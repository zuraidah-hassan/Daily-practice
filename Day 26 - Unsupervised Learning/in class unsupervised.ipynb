```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load dataset
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)

# Apply K-Means
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
df['Cluster'] = kmeans.fit_predict(df)

# Visualize clusters
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=df['Cluster'], cmap='viridis')
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.title("K-Means Clustering on Iris Dataset")
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data   # features
y = iris.target # target labels
```

```python
# Perform hierarchical clustering using Ward's method
Z = linkage(X, method='ward')
```

```python
# Create a dendrogram
plt.figure(figsize=(10, 7))
dendrogram(
    Z,
    labels=iris.target_names[y],
    leaf_rotation=90,
    leaf_font_size=12
)
plt.title('Hierarchical Clustering Dendrogram of Iris Dataset')
plt.xlabel('Iris Species')
plt.ylabel('Distance')
plt.axhline(y=7, color='r', linestyle='--')  # Optional line to indicate cut-off
plt.show()
```

```python
# Create a dendrogram # with color threshold
plt.figure(figsize=(10, 7))
dendrogram(
    Z,
    labels=iris.target_names[y],
    leaf_rotation=90,
    leaf_font_size=12,
    color_threshold=10
)
plt.title('Hierarchical Clustering Dendrogram of Iris Dataset')
plt.xlabel('Iris Species')
plt.ylabel('Distance')
plt.axhline(y=7, color='r', linestyle='--')  # Optional line to indicate cut-off
plt.show()
```
