````python
-- selection 
-- WHERE clause 

-- let's us keep rows that meet certain conditions

SELECT * from Employees;

SELECT name, salary, city, hiring_year FROM Employees WHERE city = 'New York';

-- get the employees with salary greater that 80K 

SELECT name, salary, city, hiring_year 
FROM Employees 
WHERE salary > 60000 AND city = 'Chicago' AND hiring_year = 2018;

-- AND OR 

-- USing IN 
SELECT name, salary, city, hiring_year 
FROM employees
WHERE city IN ('New York', 'Chicago');

-- USing between 
SELECT name, salary, city, hiring_year 
FROM employees
WHERE salary BETWEEN 60000 AND 800000
````

````python

#1️⃣ Drop & Create
DROP TABLE IF EXISTS employees;

CREATE TABLE employees (
    emp_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    department TEXT NOT NULL,
    salary INTEGER,
    city TEXT
);
````

````python
#2. Write INSERT statements to add at least 5 employees into the employees table with different departments, salaries, and cities.
# Insert data
INSERT INTO employees (emp_id, name, department, salary, city)
VALUES
(1, 'Ali', 'Engineering', 60000, 'Mumbai'),
(2, 'Siti', 'HR', 45000, 'Delhi'),
(3, 'Ahmad', 'Engineering', 75000, 'Bangalore'),
(4, 'Aina', 'Finance', 55000, 'Mumbai'),
(5, 'Ravi', 'Marketing', 50000, 'Chennai');
````

````python
#3. Write a query to show all columns for all employees in the table.
SELECT * FROM employees;
````

````python
#4. Write a query to show only the name and salary of all employees.
SELECT name, salary
FROM employees;
````

````python
#5. Write a query to show all employees who work in the **Engineering** department.
SELECT *
FROM employees
WHERE department = 'Engineering';
````

````python
#6. Write a query to show all employees whose salary is greater than 50,000.
SELECT *
FROM employees
WHERE salary > 50000;
````

````python
#7. Write a query to show all employees who live in the city Mumbai.
SELECT *
FROM employees
WHERE city = 'Mumbai';
````

````python
#8. Write a query to list all employees ordered by salary from highest to lowest.

SELECT *
FROM employees
ORDER BY salary DESC;
````

````python
#9. Write a query to list all employees ordered by name in A–Z order.
SELECT *
FROM employees
ORDER BY name ASC;
````

````python
#10. Write a query to show the name, salary, and a calculated column called monthly_salary which is salary / 12.
SELECT
    name,
    salary,
    salary / 12 AS monthly_salary
FROM employees;
````

````python
#11. Write an ALTER TABLE statement to add a new column called monthly_salary to the employees table.
ALTER TABLE employees
ADD COLUMN monthly_salary INTEGER;
````

````python
#12. Write an UPDATE statement to fill the monthly_salary column for all employees so that:
#monthly_salary = salary / 12.
UPDATE employees
SET monthly_salary = salary / 12;

````

````python
#13. Add another calculated column, salary_next_year, such that, the current salary has been increased by 9%
    #- Salary + 0.09 * Salary = salary_next_year
# a) add column
ALTER TABLE employees
ADD COLUMN salary_next_year INTEGER;
````
