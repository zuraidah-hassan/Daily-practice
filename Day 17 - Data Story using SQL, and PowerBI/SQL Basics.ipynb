<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>   Filtering Rows with WHERE



`WHERE` lets you keep only rows that match a condition.


````python
-- 2.1 Employees in New York
SELECT name, city, salary
FROM employees
WHERE city = 'New York';
````

````python
-- 2.2 Employees with salary greater than 70,000
SELECT name, salary
FROM employees
WHERE salary > 70000;
````

````python
-- 2.3 Employees in New York AND salary > 80,000
SELECT name, city, salary
FROM employees
WHERE city = 'New York' AND salary > 80000;
````

````python
-- 2.4 Employees in Chicago OR Boston
SELECT name, city
FROM employees
WHERE city = 'Chicago' OR city = 'Boston';
````

````python
-- 2.5 Using IN
SELECT name, city
FROM employees
WHERE city IN ('New York', 'Chicago');
````

````python
-- 2.6 Using BETWEEN
SELECT name, salary
FROM employees
WHERE salary BETWEEN 50000 AND 80000;
````

````python
-- 2.7 Using LIKE (name starts with 'A')
SELECT name, salary
FROM employees
WHERE name LIKE 'A%';
````

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>  Sorting with ORDER BY (and LIMIT)


`ORDER BY` changes the order of rows.

````python
-- 3.1 Sort by salary (low to high)
SELECT name, salary
FROM employees
ORDER BY salary ASC;
````

````python
-- 3.1 Sort by salary (low to high)
SELECT name, salary
FROM employees
ORDER BY salary ASC;
````

````python
-- 3.2 Sort by salary (high to low)
SELECT name, salary
FROM employees
ORDER BY salary DESC;
````

````python
-- 3.3 Sort by city, then salary (high â†’ low within each city)
SELECT name, city, salary
FROM employees
ORDER BY city ASC, salary DESC;
````

````python
-- 3.4 Top 3 highest paid employees (if LIMIT is supported)
SELECT name, salary
FROM employees
ORDER BY salary DESC
LIMIT 3;
````

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>  Calculated Columns & Aliases




We can create new values in the query.

````python
-- 4.1 Show monthly salary (assuming 12 months)
SELECT
    name,
    salary,
    salary / 12 AS monthly_salary
FROM employees;
````

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>  Aggregate Functions (COUNT, SUM, AVG, MIN, MAX)


````python
-- 5.1 How many employees?
SELECT COUNT(*) AS total_employees
FROM employees;
````

````python
-- 5.2 Average salary
SELECT AVG(salary) AS avg_salary
FROM employees;
````

````python
-- 5.3 Min and max salary
SELECT
    MIN(salary) AS min_salary,
    MAX(salary) AS max_salary
FROM employees;
````

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>  GROUP BY â€“ Aggregation by Category

````python
-- 6.1 Count employees per city
SELECT
    city,
    COUNT(*) AS num_employees
FROM employees
GROUP BY city;
````

````python
-- 6.2 Average salary per city
SELECT
    city,
    AVG(salary) AS avg_salary
FROM employees
GROUP BY city;
````

````python
-- 6.3 Average salary per department
SELECT
    department_id,
    AVG(salary) AS avg_salary
FROM employees
GROUP BY department_id;
````

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>  HAVING â€“ Filter Groups (After GROUP BY)

````python
-- 7.1 Cities with more than 2 employees
SELECT
    city,
    COUNT(*) AS num_employees
FROM employees
GROUP BY city
HAVING COUNT(*) > 2;
````
