<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h3>   What is a Data Story?

A **data story** is more than just charts.

It usually follows this flow:

1. **Question** – What are we trying to understand or decide?
2. **Data** – What data do we have that can help?
3. **Analysis** – Calculations, grouping, filtering, and exploring patterns.
4. **Visualization** – Charts that make patterns easy to see.
5. **Insight** – What does the data say, in simple language?
6. **Recommendation** – So what? What should we do about it?

Example (tips dataset):

> **Question:** On which days does the restaurant earn the most money?  
> **Analysis:** Sum total bill amounts per day.  
> **Visualization:** Bar chart of total revenue per day.  
> **Insight:** Friday and Saturday have the highest revenue.  
> **Recommendation:** Staff more employees on Friday and Saturday evenings.

<div style="background-color: lightgreen; color: black; padding: 4px;">
    <h3>   Plan of the steps


We will:

1. Load the `tips` dataset (restaurant bills).
2. Save it into a **SQLite** database as a table (`orders`).
3. Use **SQL queries** to answer 3 questions:
   - Q1: Which **day** brings the highest total revenue?
   - Q2: Which **day/time** has the highest **tip %**?
   - Q3: How does **party size** relate to **total bill**?
4. For each question:
   - Write a SQL query
   - Convert results to a Pandas DataFrame
   - Plot a chart
   - Write a short insight sentence

At the end, we will combine these into a **mini data story**.


````python
# 3. Setup: Import libraries

import sqlite3          # for SQLite database
import pandas as pd     # for data manipulation
import seaborn as sns   # for sample dataset
import matplotlib.pyplot as plt  # for plotting

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (7, 5)
````

### Load the Dataset (`tips`)

We use `seaborn.load_dataset("tips")`:

- Each row = one restaurant bill
- Important columns:
  - `total_bill` – total amount in dollars
  - `tip` – tip amount in dollars
  - `day` – day of week
  - `time` – Lunch or Dinner
  - `size` – number of people at the table


````python
# 4.1 Load tips dataset
tips = sns.load_dataset("tips")

# Look at first few rows
tips.head()
````

````python
# 4.2 Basic info and summary

tips.info()
````

````python
tips.describe().round(2)
````

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>   Create a SQLite Database and Save the Table

 
We will:

1. Create an **in-memory** SQLite database
2. Save the `tips` DataFrame as a table called `orders`
3. Use **SQL queries** on this table


````python
# 5.1 Create an in-memory SQLite database
conn = sqlite3.connect(":memory:")  # database lives in RAM

# 5.2 Save the tips DataFrame as a table named "orders"
tips.to_sql("orders", conn, index=False, if_exists="replace")

# 5.3 Quick check: run a simple SQL query
query = "SELECT * FROM orders LIMIT 5;"
pd.read_sql_query(query, conn)
````

#### Quick SQL Refresher (Concept)

We will use:

- `SELECT` – choose columns  
- `FROM` – specify table  
- `WHERE` – filter rows  
- `GROUP BY` – group rows to compute aggregations (sum, avg, count)  
- `ORDER BY` – sort results

Example (concept):

```sql
SELECT day, SUM(total_bill) AS total_revenue
FROM orders
GROUP BY day
ORDER BY total_revenue DESC;


This means:

Group bills by day

Sum the total_bill for each day

Sort days by total revenue (highest first)

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>   PART A – Question 1: Which day brings the highest total revenue?

We want to know:

> On which **day** does the restaurant make the **most total money**?

Steps:

1. Use SQL: group by `day`, sum `total_bill`
2. Sort by total revenue (descending)
3. Visualize with a bar chart
4. Write a simple insight


````python
# A.1 SQL query: total revenue by day

query_q1 = """
SELECT
    day,
    SUM(total_bill) AS total_revenue
FROM orders
GROUP BY day
ORDER BY total_revenue DESC;
"""

revenue_by_day = pd.read_sql_query(query_q1, conn)
revenue_by_day
````

````python
# A.2 Plot total revenue by day

plt.figure()
sns.barplot(data=revenue_by_day, x="day", y="total_revenue")
plt.title("Total Revenue by Day")
plt.xlabel("Day of Week")
plt.ylabel("Total Revenue ($)")
plt.show()
````

#### Insight for Question 1

You can guide students to write something like:

> **Insight Q1:**  
> The restaurant earns the highest total revenue on **[top day]**, followed by **[second day]**.  
> This suggests that these days are the **busiest** or **highest spending** days.

In class, ask them:

- Why might that day be higher? (e.g., weekend dining)
- What could the restaurant do on those days? (e.g., more staff, promotions)


<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>   PART B – Question 2: On which day/time is tip percentage highest?


Here we ask:

> When do customers leave the **best tips** as a percentage of the bill?

We will:

1. Add a **tip percentage** = tip / total_bill * 100
2. Use SQL to compute **average tip %** grouped by `day` and `time`
3. Visualize using a bar plot
4. Write an insight


````python
# B.1 Add tip percentage as a column in the SQL table as well

# First, we can add it in Pandas
tips["tip_pct"] = tips["tip"] / tips["total_bill"] * 100

# Write updated table back to SQLite (replace old one)
tips.to_sql("orders", conn, index=False, if_exists="replace")

# Quick check
pd.read_sql_query("SELECT tip, total_bill, tip_pct FROM orders LIMIT 5;", conn)
````

````python
# B.2 SQL query: average tip % by day and time

query_q2 = """
SELECT
    day,
    time,
    AVG(tip_pct) AS avg_tip_pct
FROM orders
GROUP BY day, time
ORDER BY avg_tip_pct DESC;
"""

tip_pct_by_day_time = pd.read_sql_query(query_q2, conn)
tip_pct_by_day_time
````

````python
# B.3 Plot average tip % by day and time

plt.figure()
sns.barplot(
    data=tip_pct_by_day_time,
    x="day",
    y="avg_tip_pct",
    hue="time"
)
plt.title("Average Tip % by Day and Time")
plt.xlabel("Day of Week")
plt.ylabel("Average Tip %")
plt.legend(title="Time")
plt.show()
````

#### Insight for Question 2

Sample guided insight:

> **Insight Q2:**  
> The highest average tip percentage occurs on **[day/time combination]**.  
> This suggests that customers are more generous with tips during **[e.g., weekend dinners]**.  
> The restaurant might focus on **excellent service** and **premium offerings** during these periods.

Ask students:

- Do they see any differences between Lunch and Dinner?
- What might explain these patterns? (e.g., mood, type of customer, special occasions)


<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4>   PART C – Question 3: Does party size affect total bill?



Now we ask:

> Do larger groups tend to spend more?

We will:

1. Use SQL to compute **average total_bill per party size**
2. Visualize using a bar chart or line
3. Write an insight in simple language


````python
# C.1 SQL query: average total_bill by size

query_q3 = """
SELECT
    size,
    AVG(total_bill) AS avg_total_bill,
    COUNT(*) AS num_orders
FROM orders
GROUP BY size
ORDER BY size;
"""

bill_by_size = pd.read_sql_query(query_q3, conn)
bill_by_size
````

````python
# C.2 Plot average total bill by party size

plt.figure()
sns.barplot(data=bill_by_size, x="size", y="avg_total_bill")
plt.title("Average Total Bill by Party Size")
plt.xlabel("Party Size (Number of People)")
plt.ylabel("Average Total Bill ($)")
plt.show()
````

#### Insight for Question 3

Example insight:

> **Insight Q3:**  
> Larger groups (higher party size) tend to have a **higher average bill**.  
> This is expected, since more people usually order more items.  
> The restaurant can plan seating and staff based on the most common group sizes.

You can also point out:

- `num_orders` column tells us how many tables had each size.
- If one size has very few observations, we should be cautious about the conclusion.
