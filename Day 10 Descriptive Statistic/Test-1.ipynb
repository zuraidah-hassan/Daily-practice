# Bubble Sort menyusun list atau array daripada kecil ke besar (ascending) atau besar ke kecil (descending).
# belajar cara swap, loop, dan logik programming.

```python
# Contoh bubble sort tanpa guna .sort() atau method lain:
numbers = [6, 4, 1, 5, 2]

for i in range(len(numbers)):
    for j in range(len(numbers)-1):
        if numbers[j] > numbers[j+1]:
            temp = numbers[j]
            numbers[j] = numbers[j+1]
            numbers[j+1] = temp

print('Sorted:', numbers)
```

```python
# Bubble Sort — hanya print selepas setiap pass

numbers = [6, 4, 1, 5, 2]

print("Original list:", numbers)

for i in range(len(numbers)):
    for j in range(len(numbers)-1):
        if numbers[j] > numbers[j+1]:
            # swap manual
            temp = numbers[j]
            numbers[j] = numbers[j+1]
            numbers[j+1] = temp
    print(f"After pass {i+1}: {numbers}")  # print selepas satu pass selesai

print("Sorted list:", numbers)
```

```python
#Bubble Sort Manual
def manual_bubble_sort(arr):
    # buat salinan supaya original tak berubah
    arr_copy = arr.copy()
    for i in range(len(arr_copy)):
        for j in range(len(arr_copy)-1):
            if arr_copy[j] > arr_copy[j+1]:
                temp = arr_copy[j]
                arr_copy[j] = arr_copy[j+1]
                arr_copy[j+1] = temp
    return arr_copy

my_list = [7, 3, 5, 2, 4]
sorted_list = manual_bubble_sort(my_list)

print("Original:", my_list)   # masih [7,3,5,2,4]
print("Sorted:", sorted_list) # [2,3,4,5,7]

```

# Selection Sort
* susun list dari kecil ke besar (ascending)
1.Cari elemen terkecil dalam list yang belum disusun.
2.Tukar elemen itu dengan elemen pada posisi pertama yang belum disusun.
3.Pindah ke elemen seterusnya dan ulang langkah 1-2 sehingga list selesai.


```python
def selection_sort(arr):
    arr_copy = arr.copy()  # salin supaya original tak berubah
    n = len(arr_copy)
    
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr_copy[j] < arr_copy[min_index]:
                min_index = j
        # swap manual
        if min_index != i:
            temp = arr_copy[i]
            arr_copy[i] = arr_copy[min_index]
            arr_copy[min_index] = temp
    
    return arr_copy

# contoh penggunaan
my_list = [7, 3, 5, 2, 4]
sorted_list = selection_sort(my_list)
print("Original:", my_list)
print("Sorted:", sorted_list)
```

```python
def selection_sort_step(arr):
    arr_copy = arr.copy()  # salin supaya original tak berubah
    n = len(arr_copy)
    
    print("Original list:", arr_copy)
    
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr_copy[j] < arr_copy[min_index]:
                min_index = j
        if min_index != i:
            # swap manual
            temp = arr_copy[i]
            arr_copy[i] = arr_copy[min_index]
            arr_copy[min_index] = temp
            print(f"Swap index {i} and {min_index}: {arr_copy}")  # print setiap swap
        else:
            print(f"No swap needed for pass {i+1}: {arr_copy}")
    
    print("Sorted list:", arr_copy)
    return arr_copy

# contoh penggunaan
my_list = [7, 3, 5, 2, 4]
selection_sort_step(my_list)
```

```python
#universal step by step
#pilih Bubble Sort atau Selection Sort, dan program akan tunjuk setiap swap / setiap pass.

def bubble_sort_step(arr):
    arr_copy = arr.copy()
    print("Original list:", arr_copy)
    
    for i in range(len(arr_copy)):
        for j in range(len(arr_copy)-1):
            if arr_copy[j] > arr_copy[j+1]:
                arr_copy[j], arr_copy[j+1] = arr_copy[j+1], arr_copy[j]
                print(f"Swap index {j} and {j+1}: {arr_copy}")
        print(f"After pass {i+1}: {arr_copy}")
    print("Sorted list:", arr_copy)
    return arr_copy

def selection_sort_step(arr):
    arr_copy = arr.copy()
    n = len(arr_copy)
    print("Original list:", arr_copy)
    
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr_copy[j] < arr_copy[min_index]:
                min_index = j
        if min_index != i:
            arr_copy[i], arr_copy[min_index] = arr_copy[min_index], arr_copy[i]
            print(f"Swap index {i} and {min_index}: {arr_copy}")
        else:
            print(f"No swap needed for pass {i+1}: {arr_copy}")
    print("Sorted list:", arr_copy)
    return arr_copy

# input user
user_input = input("Masukkan nombor dipisahkan dengan koma: ")  # contoh: 7,3,5,2,4
user_list = [int(x.strip()) for x in user_input.split(",")]

# pilih algoritma
alg = input("Pilih algoritma (bubble/selection): ").strip().lower()

if alg == "bubble":
    bubble_sort_step(user_list)
elif alg == "selection":
    selection_sort_step(user_list)
else:
    print("Algoritma tidak dikenali. Gunakan 'bubble' atau 'selection'.")
```

```python
#Insertion Sort (Manual)
*Susun list secara bertahap.
*Ambil satu elemen dari list tidak tersusun.
*Letakkan di posisi yang betul dalam list yang sudah tersusun.
```

```python
def insertion_sort(arr):
    arr_copy = arr.copy()
    for i in range(1, len(arr_copy)):
        key = arr_copy[i]
        j = i - 1
        # pindahkan elemen yang lebih besar dari key ke kanan
        while j >= 0 and arr_copy[j] > key:
            arr_copy[j+1] = arr_copy[j]
            j -= 1
        arr_copy[j+1] = key
    return arr_copy

# penggunaan
my_list = [7, 3, 5, 2, 4]
sorted_list = insertion_sort(my_list)
print("Original:", my_list)
print("Sorted:", sorted_list)
```

```python
# Manual Sorting dengan Min/Max Loop (Simple)
# buat list baru, cari elemen terkecil (atau terbesar) berulang kali, tambah ke list baru

def manual_min_sort(arr):
    arr_copy = arr.copy()
    sorted_list = []
    while arr_copy:
        min_val = arr_copy[0]
        for x in arr_copy:
            if x < min_val:
                min_val = x
        sorted_list.append(min_val)
        arr_copy.remove(min_val)
    return sorted_list

my_list = [7, 3, 5, 2, 4]
print("Sorted:", manual_min_sort(my_list))
```

*Bubble Sort → bandingkan dan swap bersebelahan.
*Selection Sort → cari elemen terkecil dan swap.
*Insertion Sort → masukkan elemen pada posisi betul dalam list yang sudah tersusun.
*sesuai untuk belajar logik loop & swap.
