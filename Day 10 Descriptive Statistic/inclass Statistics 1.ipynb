```python
nums = [20,80,130,50,90,89]
nums.sort()
print('sorted list nums: ',nums)
# if the nums is odd, length/2
# 3/2 = 1.5
# 7/2 = 3.5
#  0  1   2   3    4
# [5, 20, 50, 80, 130]

mean = sum(nums) / len(nums)

print('mean of the nums is: ',mean)

if len(nums)%2 == 0:
    #if length is even
    median = (nums[len(nums)//2] + nums[(len(nums)//2) - 1]) / 2
    print("median is: ",median)
else:
    # if the length is odd
    median = nums[len(nums)//2]
    print("median is: ",median)


# nums[2]
```

```python
 # variance
import numpy as np

class1 = np.array([50,47,51,49,47])

mean = class1.mean()
print("Mean is:", mean)

# kira variance
sum_of_diff = (class1 - mean)**2
variance = sum_of_diff.sum() / len(class1)

print("The variance is:", variance)
```

```python
# Median = Sort the numbers
# Median ialah nilai tengah dalam satu set data apabila data disusun dari yang terkecil ke yang terbesar.
# Ia menunjukkan “tengah-tengah” data.

data = [50, 47, 51, 49, 47]

data.sort()
print("Sorted data:", data)
```

```python
# Targeted the middle numbers
n = len(data)

if n % 2 == 1:  # n ganjil
    median = data[n // 2]
else:           # n genap
    median = (data[n//2 - 1] + data[n//2]) / 2

print("Median:", median)
```
