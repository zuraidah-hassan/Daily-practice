```python
#bar plot
plt.figure(figsize=(8,5))
sns.barplot(
    data=df,
    x='species',
    y='body_mass_g',
    palette='viridis'
)
plt.title("Average Body Mass by Penguin Species", fontsize=14, weight='bold')
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(8,5))
sns.scatterplot(
    data=df,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    style='sex',
    s=80
)
plt.title("Bill Length vs Bill Depth", fontsize=14, weight='bold')
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()
```

```python
#pairplot
sns.pairplot(df, hue='species')
```

```python
#all 3 subplot chart
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset("penguins")

# Clean missing values (buang NaN)
df = df.dropna()

# Create figure
plt.figure(figsize=(16,5))

# ---------------------------
# 1 Bar Plot
# ---------------------------
plt.subplot(1, 3, 1)
sns.barplot(
    data=df,
    x='species',
    y='body_mass_g',
    palette='pastel'
)
plt.title("Average Body Mass by Species")
plt.xticks(rotation=20)

# ---------------------------
# 2 Scatter Plot
# ---------------------------
plt.subplot(1, 3, 2)
sns.scatterplot(
    data=df,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    s=70
)
plt.title("Bill Length vs Bill Depth")

# ---------------------------
# 3 Histogram
# ---------------------------
plt.subplot(1, 3, 3)
sns.histplot(
    data=df,
    x='flipper_length_mm',
    kde=True,
    color='purple'
)
plt.title("Flipper Length Distribution")

plt.tight_layout()
plt.show()
```

```python
# tips dataset

df = sns.load_dataset('tips') #tips.csv
```

```python
df.head()
```

```python
df.shape
```

```python
# create bargraph day vs tips
```

```python
#bargraph
plt.figure(figsize=(8,5))
sns.barplot(
    data=df,
    x='day',
    y='tip',             # nilai tip
    estimator=sum,       # jumlah tip untuk setiap hari
    palette='coolwarm'
)
plt.title("Total Tips by Day", fontsize=14, weight='bold')
plt.xlabel("Day")
plt.ylabel("Total Tip ($)")
plt.show()
```

```python
#pairplot
sns.pairplot(df, hue='sex', height=2.5, palette='pastel')
plt.show()
```

```python
#histogram
plt.figure(figsize=(8,5))
sns.histplot(
    data=df,
    x='total_bill',     # column nak plot
    bins=20,            # jumlah bins
    kde=True,           # overlay KDE (density curve)
    color='skyblue'
)
plt.title("Distribution of Total Bill", fontsize=14, weight='bold')
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.show()
```
