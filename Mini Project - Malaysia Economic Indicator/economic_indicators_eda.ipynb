```python
df = pd.read_csv("malaysian_economic.csv")
```

```python
df.head()
df.info()
df.describe()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.dtypes
```

```python
#FIGURE 1ï¸âƒ£ â€“ Overall Economic Cycle (Plotly)
import plotly.express as px

fig1 = px.line(
    df,
    x='date',
    y=['leading', 'coincident', 'lagging'],
    title="Malaysia Economic Indicators Across Business Cycles",
    labels={'value': 'Index Value', 'variable': 'Indicator'},
    template='plotly_dark'
)

fig1.show()
```

# Visual 1
The leading indicator exhibits earlier turning points compared to coincident and lagging indicators, supporting its role in anticipating changes in Malaysiaâ€™s economic cycle.

```python
#FIGURE 2ï¸âƒ£ â€“ Diffusion Index (Momentum Analysis)
fig2 = px.line(
    df,
    x='date',
    y=['leading_diffusion', 'coincident_diffusion'],
    title="Diffusion Index: Economic Momentum",
    labels={'value': 'Percentage of Expanding Sectors (%)'},
    template='plotly_dark'
)

fig2.show()
```

# Visual 2
The diffusion index shows periods of economic expansion and contraction. Movements in the leading diffusion index generally precede changes in the coincident diffusion index, indicating early shifts in economic momentum.

```python
#Figure 3 (Scatter Plot â€“ Leading vs Coincident)
fig3 = px.scatter(
    df,
    x='leading',
    y='coincident',
    color=df['leading_diffusion'] > 50,
    title="Leading vs Coincident Indicators by Economic Regime",
    labels={'color': 'Expansion Regime'},
    template='plotly_dark'
)

fig3.show()
```

# Visual 3
The scatter plot highlights a strong positive association between leading and coincident indicators, with expansion regimes clustered at higher indicator values.

```python
#Figure 4 (Box Plot
import plotly.express as px

import plotly.express as px

df_long = df.melt(
    value_vars=['leading', 'coincident', 'lagging'],
    var_name='Indicator',
    value_name='Value'
)

fig4 = px.box(
    df_long,
    x='Indicator',
    y='Value',
    title="Distribution of Leading, Coincident, and Lagging Economic Indicators",
    template='plotly_dark'
)

fig4.show()
```

# Visual 4
The box plot highlights differences in variability across indicators, with lagging indicators displaying wider dispersion than leading indicators.

```python
fig4 = px.parallel_coordinates(
    df,
    dimensions=[
        'leading', 'coincident', 'lagging',
        'leading_diffusion', 'coincident_diffusion'
    ],
    title="Parallel Coordinates of Malaysiaâ€™s Economic Indicators",
    template='plotly_dark'
)

fig4.show()
```

```python
fig4 = px.histogram(
    df_long,
    x='Value',
    facet_col='Indicator',
    nbins=40,
    title="Distribution of Economic Indicators",
    template='plotly_dark'
)

fig4.show()
```

# Conclusion & Key Insights

* Leading indicators provide early signals of changes in economic activity.

* Diffusion indices reflect economic momentum, with leading diffusion moving ahead of coincident diffusion.

* Lagging indicators are more volatile and mainly serve to confirm economic cycles.
