```python
import numpy as np

# Matriks A 3x3
A = np.array([
    [2, 3, 4],
    [6, 7, 8],
    [3, 2, 3]
])

# Matriks B 2x3
B = np.array([
    [5, 5, 5],
    [2, 2, 2]
])

B @ A
# matrix multiplication is not commutative
```

```python
#Darab matriks (matrix multiplication)
# Darab B dan A
C = np.dot(B, A)

print("B · A =")
print(C)
```

```python
A = np.array([
    [5, 5, 5],
    [2, 2, 2],
    [5, 5, 5]
])

# Matriks B 2x3
B = np.array([
    [3, 3, 3],
    [3, 3, 3],
    [3, 3, 3]
])

print ('Matrix Multiplication: \n',A @ B,'\n')
print('Array Multiplication: \n',A * B)
```
