    <h3> Why Linear Algebra in Machine Learning?

Machine Learning models often work with data like:

- Features of a house: `[size, number_of_rooms, age]`
- Features of a person: `[height, weight, age]`
- Pixels of an image: `[pixel_1, pixel_2, ..., pixel_n]`

We usually store these as **vectors** and **matrices**.

Examples:

- One person’s features → a **vector**
- Many people’s features → a **matrix** (each row is one person)

Linear algebra gives us the tools to:

- Store many numbers together
- Transform data
- Combine features with **weights** (like in Linear Regression and Neural Networks)

You do **not** need to become a math expert today.  
You just need to be **comfortable** with the basic operations.

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Scalars, Vectors, and Matrices

#### Scalar

A **scalar** is just a single number.

Examples:
- 3  
- -5.2  
- 100  

In ML, scalars could be:
- A single weight
- A learning rate
- A single feature value

```python
# Import NumPy and quick setup

import numpy as np

# Just to see printed arrays more clearly
np.set_printoptions(suppress=True)  # don't use scientific notation for small numbers
```

```python
# Scalar: just a number in Python
scalar = 5
print("Scalar:", scalar)
```

#### Vector

A **vector** is an ordered list of numbers.  
You can think of it as a **1D array**.

Example of a vector with 3 elements:

x =
\begin{bmatrix}
2 \\
5 \\
7
\end{bmatrix}


In code, we might write: `x = [2, 5, 7]`

Examples in real life:
- A student’s marks in 3 subjects: `[80, 75, 90]`
- Weather features: `[temperature, humidity, wind_speed]`

```python
# Vector: use np.array with 1D list
vector = np.array([2, 4, 6])
print("Vector:", vector)
print("Vector shape:", vector.shape)  # (3,) means 1D with 3 elements
```

#### Matrix

A **matrix** is a rectangular table of numbers → rows and columns.

Example:


A =
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{bmatrix}

- This matrix has **2 rows** and **3 columns**
- We say its **shape** is `(2, 3)`

In real life:
- A table of students and marks

| Student | Math | Science | English |
|---------|------|---------|---------|
| A       | 80   | 75      | 90      |
| B       | 60   | 85      | 70      |

If we drop the "Student" names and keep only numbers, we get a matrix:

Marks =
\begin{bmatrix}
80 & 75 & 90 \\
60 & 85 & 70
\end{bmatrix}


```python
# Matrix: use np.array with nested lists
matrix = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
print("\nMatrix:\n", matrix)
print("Matrix shape:", matrix.shape)  # (2, 3) means 2 rows, 3 columns
```

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Vector Addition

If we have two vectors of the **same length**, we can add them **element by element**.

Example:

a =
\begin{bmatrix}
1 \\
3 \\
5
\end{bmatrix},


b=
\begin{bmatrix}
2 \\
4 \\
6
\end{bmatrix}

Then:

a + b = 

\begin{bmatrix}
1+2 \\
3+4 \\
5+6
\end{bmatrix}

=

\begin{bmatrix}
3 \\
7 \\
11
\end{bmatrix}

We’re just adding the corresponding positions.

```python
# Create two vectors of the same length
a = np.array([1, 3, 5])
b = np.array([2, 4, 6])

print("Vector a:", a)
print("Vector b:", b)

# Vector addition
c = a + b
print("a + b =", c)
```

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Scalar Multiplication

We can multiply a vector by a **scalar** (single number) by multiplying each element.

Example:

a =
\begin{bmatrix}
2 \\
4 \\
6
\end{bmatrix},


k = 3


Then:


k . a =

\begin{bmatrix}
3 \cdot 2 \\
3 \cdot 4 \\
3 \cdot 6
\end{bmatrix}
=
\begin{bmatrix}
6 \\
12 \\
18
\end{bmatrix}

This is like scaling or stretching the vector.

```python
# Scalar (single number)
k = 3

# Vector from before: a = [1, 3, 5]
print("Vector a:", a)
print("Scalar k:", k)

# Multiply each element of vector by the scalar
k_times_a = k * a
print("k * a =", k_times_a)
```

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Dot Product

The **dot product** of two vectors of the same length is:

1. Multiply corresponding elements
2. Add up the results

Example:

a =
\begin{bmatrix}
1 \\
2 \\
3
\end{bmatrix},

b =
\begin{bmatrix}
4 \\
5 \\
6
\end{bmatrix}


Then:

a . b = 1 . 4 + 2 . 5 + 3 . 6 
= 4 + 10 + 18 
= 32



#### Why is this useful?

Think of a **simple model**:


score = w_1 . x_1 + w_2 . x_2 + w_3 . x_3


- \(x_1, x_2, x_3\) → features (like [hours_studied, sleep_hours, coffee_cups])  
- \(w_1, w_2, w_3\) → weights learned by the model

This formula is exactly a **dot product** between:


- w = [w_1, w_2, w_3] , x = [x_1, x_2, x_3]


```python
# Define two vectors for dot product
u = np.array([1, 2, 3])
v = np.array([4, 5, 6])

print("u:", u)
print("v:", v)

# Method 1: using np.dot
dot1 = np.dot(u, v)

# Method 2: using @ operator (matrix multiplication operator)
dot2 = u @ v

print("Dot product (np.dot):", dot1)
print("Dot product (u @ v):", dot2)
```

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> Matrix-Vector Multiplication

Suppose we have:

- A matrix **A** of shape (2, 3)

A =
\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{bmatrix}

- A vector **x** of length 3

x =
\begin{bmatrix}
1 \\
0 \\
-1
\end{bmatrix}


Then the product A and x gives a **new vector** of length 2:


A * x =
\begin{bmatrix}
1\cdot1 + 2\cdot0 + 3\cdot(-1) \\
4\cdot1 + 5\cdot0 + 6\cdot(-1)
\end{bmatrix}

=

\begin{bmatrix}
1 + 0 - 3 \\
4 + 0 - 6
\end{bmatrix}
=
\begin{bmatrix}
-2 \\
-2
\end{bmatrix}


You can think of it as:

- Each **row** of the matrix does a **dot product** with the vector.
- This appears a lot in ML when:
  - You have many samples (rows)
  - Each sample is multiplied by the same weight vector

```python
# Matrix A with shape (2, 3)
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

# Vector x with length 3
x = np.array([1, 0, -1])

print("Matrix A:\n", A)
print("Vector x:", x)
print("A shape:", A.shape)
print("x shape:", x.shape)

# Matrix-vector multiplication using @
Ax = A @ x
print("\nA @ x =", Ax)
```

<div style="background-color: lightblue; color: black; padding: 4px;">
    <h4> A Simple ML-Style Example: Weighted Sum of Features

Imagine you have a student with 3 features:

- `hours_studied`
- `hours_sleep`
- `social_media_hours` (per day)

You have a simple model that computes a “score”

```python
# Features of one student: [hours_studied, hours_sleep, social_media_hours]
x_features = np.array([4, 7, 2])

# Weights learned by some model (just an example)
w_weights = np.array([3.0, 1.0, -2.0])

print("Features (x):", x_features)
print("Weights (w): ", w_weights)

# Score = w · x
score = w_weights @ x_features

print("\nModel score (w · x) =", score)
```

#### Interpretation

- `3.0 * hours_studied` → studying more increases the score  
- `1.0 * hours_sleep` → sleeping more slightly increases the score  
- `-2.0 * social_media_hours` → more social media decreases the score  

This **dot product** pattern appears again and again in ML models.
