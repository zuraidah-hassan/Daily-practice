```python
# data

# bar graph

# create a list of students
# grades


students = ['Ida', 'Amy', 'Bell', 'Kimi', 'Max', 'Leo']
grades = [75, 95, 87, 71, 97, 68]

cities = ["KL","SG","Jakarta","Bangkok","Manila"]
temp = [28,45,13,35,7]

```

```python
# import matplotlib
import matplotlib.pyplot as plt

# create a bar graph
plt.bar(students, grades, color = "green") #specify properties of our bar chart

plt.title("Chart of Students grades")

plt.xlabel("student names")

plt.ylabel("marks")

plt.show() #display the bar chart
```

```python
# create a  scatter plot
plt.scatter(students, grades, color = "green") #specify properties of our bar chart

# prettying up the visual
plt.title("Chart of Students grades")
plt.xlabel("student names")
plt.ylabel("marks")
plt.show() #display the scatter chart
```

# create a bar chart for cities and temperature

```python
# create a sactter plot
plt.figure(figsize=(3,3))
plt.scatter(students, grades, color = 'green') # specify properties of our bar chart

# prettying up the visual
plt.title('Chart of student grades')
plt.xlabel('student names')
plt.ylabel('marks')
plt.show() 
```

```python
# create a figure of width and height and place 3 plots side-by-side
plt.figure(figsize=(15,5))   # define the size  

# subplot 1 - bar chart
plt.subplot(1,3,1)  # rows, columns, position
plt.bar(students, grades, color='green')  # bar chart
plt.title('Bar: Student Grades')
plt.xlabel('student names')
plt.ylabel('marks')

# subplot 2 - line chart
plt.subplot(1,3,2)
plt.plot(students, grades, color='green', marker='o')  # line chart
plt.title('Line: Student Grades')
plt.xlabel('student names')
plt.ylabel('marks')

# subplot 3 - scatter chart
plt.subplot(1,3,3)
plt.scatter(students, grades, color='green')  # scatter chart
plt.title('Scatter: Student Grades')
plt.xlabel('student names')
plt.ylabel('marks')

plt.tight_layout()
plt.show()
```
